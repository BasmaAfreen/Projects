# 	Boxplots:
####              1. Create boxplots for numerical variables to identify outliers and the interquartile range.
         


```python
 #**sns**
numeric_col=df.select_dtypes(include=['number']).columns
for col in numeric_col:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
numeric_col=df.select_dtypes(include=['number']).columns

# Identify columns with outliers using boxplots (replace with your actual outlier detection logic if needed)
# Initialize an empty list to store columns that contain outliers
outlier_cols = []

# Loop through each numerical column to detect outliers
for col in numeric_col:
   # Calculate the first quartile (Q1)
    Q1 = df[col].quantile(0.25)
    # Calculate the third quartile (Q3)
    Q3 = df[col].quantile(0.75)
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    # Define the lower bound for outliers
    lower_bound = Q1 - 1.5 * IQR
    # Define the upper bound for outliers
    upper_bound = Q3 + 1.5 * IQR
    # Check if there are any outliers in the column
    if df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0] > 0 :
    # If outliers are found, append the column name to outlier_cols
      outlier_cols.append(col)

print(df.describe())

# Print the columns that contain outliers
print("Outlier Columns:", outlier_cols)

# Replace outliers with the median for each identified column
for col in outlier_cols:
    # Calculate the median value of the column
    median_val = df[col].median()
    # Recalculate Q1 and Q3 for accuracy
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Replace outliers with the median value
    df.loc[(df[col] < lower_bound) | (df[col] > upper_bound), col] = median_val
```

```python
for col in numeric_col:
    plt.figure(figsize=(6, 4))
    plt.boxplot(x=df[col],vert=False)
    plt.xlabel(col)
    plt.title(f"Boxplot  {col} ")
    plt.show()
```

# Conversion of Categorical Data into Dummy Variables
-	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
-   Steps:
-	Discuss the need for converting categorical data into dummy variables (one-hot encoding).



-	            Converting categorical data into dummy variables (one-hot encoding) is important for several reasons:

1. **Machine Understanding**: Machines can't understand wordThey need numerical data to process and analyze,s, so we convert categories to number
2. **Library Requirements** : Many machine learning libraries, like scikit-learn, do not accept or process strings as inputs for training models. They                               require numerical input data. One-hot encoding ensures that our categorical data is in the right format for these                                         libraries to handle.
3. **Avoiding Misinterpretation**: One-hot encoding prevents the model from thinking categories have an order.
4. **Preserving Information**: Each category gets its own column, ensuring no information is lost.
5. **Better Model Performance**: Proper encoding helps the model learn accurately and improves perform


dtaset.


###  Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.

```python
#Extract Categorical Variables:
categorical_variables=df.loc[: ,["Day","SKU","City","BU","Brand","Model"]]

# unique() method is applied on single column or series
#Get and sort unique values for each column.
A=sorted(categorical_variables["Day"].unique())
B=sorted(categorical_variables["SKU"].unique())
C=sorted(categorical_variables["City"].unique())
D=sorted(categorical_variables["BU"].unique())
E=sorted(categorical_variables["Brand"].unique())
F=sorted(categorical_variables["Model"].unique())

#columns is a concatenated list of sorted unique values, matching the number of columns in your encoded DataFrame
#Concatenate Unique Values into Column Names
col=A+B+C+D+E+F

# Note: The line j4.columns = [b6, b7] will cause an issue since you're trying to set the DataFrame columns with two lists, which doesn't match the
#total number of encoded columns ,SO concatenate the sorted unique values
```

```python
#Use the OneHotEncoder to transform categorical data
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
Dummy_col= OHE.fit_transform(categorical_variables).toarray()

#Convert the encoded array to a DataFrame
Dummy_col=pd.DataFrame(Dummy_col)

#set the columns.
Dummy_col.columns=col

```
